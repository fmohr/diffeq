## Three Tanks

Considerer the linear system
$$ \dfrac{dx_1}{dt}=-\dfrac{3}{50}x_1+\dfrac{1}{50}x_2$$
$$ \dfrac{dx_2}{dt}=\dfrac{1}{25}x_1-\dfrac{1}{20}x_2+\dfrac{1}{100}x_3$$
$$ \dfrac{dx_3}{dt}=\dfrac{1}{20}x_2-\dfrac{1}{20}x_3 $$
with $x_1(0)=x_2(0)=x_3(0)=50$.


### Explicit solution
$$x_1(t)=88.85294(0.307e^{-0.01659 t})+55.937(0.2428e^{-0.05377 t})+17.0844(0.53502e^{-0.08962 t})$$
$$x_2(t)=88.85294(0.668e^{-0.01659 t})+55.937(0.0755e^{-0.05377 t})-17.0844(0.7947e^{-0.08962 t})$$
$$x_3(t)=88.85294(e^{-0.01659 t})-55.937(e^{-0.05377 t})+17.0844(e^{-0.08962 t})$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# time points
t = np.linspace(0,300,30)
x_1 = 88.85294*(0.307*np.exp(-0.01659*t))+55.937*(0.2428*np.exp(-0.05377*t))+17.0844*(0.53502*np.exp(-0.08962*t))
x_2 = 88.85294*(0.668*np.exp(-0.01659*t))+55.937*(0.0755*np.exp(-0.05377*t))-17.0844*(0.7947*np.exp(-0.08962*t))
x_3 = 88.85294*(np.exp(-0.01659*t))-55.937*(np.exp(-0.05377*t))+17.0844*(np.exp(-0.08962*t))
# plot results
plt.plot(t,x_1,'b-',label=r'$x_1(t)$')
plt.plot(t,x_2,'r--',label=r'$x_2(t)$')
plt.plot(t,x_3,'b--',label=r'$x_3(t)$')
plt.ylabel('real solution')
plt.xlabel('time')
plt.show()
```

```python
step_size = 30
t = np.linspace(0,300,step_size)
gen = ThreeTankDataGenerator()
obs = gen.generate_orig(t)
```

```python
print(obs)
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 4))
ax1.plot(t, obs[:,0], 'o')
ax2.plot(t, obs[:,1], '*')
ax3.plot(t, obs[:,2], '+')
plt.show()
```

```python
evaluator = ODEINTEvaluator(t, obs)
```

```python
print("Final Error:", evaluator.evaluate([lambda x1, x2, x3, t, a, b, c: a*x1 + b*x2 + c*x3, lambda x1, x2, x3, t, d, e, f: d*x1 + e*x2 + f*x3, lambda x1, x2, x3, t, g, h, i: g*x1 + h*x2 + i*x3 ],attempts_per_exponent=1000))
```
