```python
from lmfit import minimize, Parameters, Parameter, report_fit
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(xs, t, ps):
    try:
        a = ps['a'].value
        b = ps['b'].value
    except:
        a, b = ps
    x = xs

    return np.sin(a*x) + b*np.sqrt(x)

def g(t, x0, ps):
    #Solution to the ODE x'(t) = f(t,x,k) with initial condition x(0) = x0
    x = odeint(f, x0, t, args=(ps,))
    return x

def residual(ps, ts, data):
    x0 = ps['x0'].value
    model = g(ts, x0, ps)
    return (model - data).ravel()
```

```python
a = 4
b = 1
true_params = [a, b]
x0 = 0.1

t = np.linspace(0, 5, 20)
data = g(t, x0, true_params)
for i in range(10):
    # set parameters incluing bounds
    a0 = 10*(np.random.rand()-0.5)
    b0 = 10*(np.random.rand()-0.5)
    params = Parameters()
    params.add('x0', value=float(data[0]), min=0, max=100)
    params.add('a', value= a0, min=-10, max=10)
    params.add('b', value= b0, min=0, max=3) #Parametro sensible

    # fit model and find predicted values
    result = minimize(residual, params, args=(t, data), method='leastsq',max_nfev = 10000)
    final = data + result.residual.reshape(data.shape)

    # plot data and fitted curves
    plt.plot(t, data, 'o')
    plt.plot(t, final, '--', linewidth=2, c='blue');
    plt.show()
    print("Error is:" + str(result.chisqr))
    print("a initial:" + str(a0))
    print("b initial:" + str(b0))
    report_fit(result)
```
